```python
def replace_outliers(feedback_modality):
    # Calculate mean time excluding outliers
    valid_times = [item['time_to_complete_minutes'] for item in feedback_modality if item['time_to_complete_minutes'] <= 100]
    if valid_times:  # Check if the list is not empty
        mean_time = sum(valid_times) / len(valid_times)
    else:
        mean_time = 0  # Default mean time if all values are outliers

    # Replace values over 100 with the mean time
    for item in feedback_modality:
        if item['time_to_complete_minutes'] > 100:
            item['time_to_complete_minutes'] = mean_time
    
    return feedback_modality


# Update each modality group by reassigning the returned updated list
specific_feedback_modality = replace_outliers(specific_feedback_modality)
general_feedback_modality = replace_outliers(general_feedback_modality)
control_group = replace_outliers(control_group)
```

## cardiffnlp/twitter-roberta-base-sentiment-latest

```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer, AutoConfig
import torch
from scipy.special import softmax
import numpy as np

# Initialize tokenizer and model
MODEL = "cardiffnlp/twitter-roberta-base-sentiment-latest" # english -> use english data for this 
tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = AutoModelForSequenceClassification.from_pretrained(MODEL)

def preprocess(text):
    new_text = []
    for t in text.split(" "):
        t = '@user' if t.startswith('@') and len(t) > 1 else t
        t = 'http' if t.startswith('http') else t
        new_text.append(t)
    return " ".join(new_text)

def chunk_text(text, max_length):
    # Tokenize the text to get a list of tokens
    tokens = tokenizer.tokenize(preprocess(text))
    # Calculate the number of chunks needed
    num_chunks = len(tokens) // (max_length - 2) + 1  # subtract 2 for special tokens [CLS] and [SEP]
    # Split the tokens into roughly equal parts
    chunks = [
        tokens[i * (max_length - 2): (i + 1) * (max_length - 2)]
        for i in range(num_chunks)
    ]
    return chunks

def predict_sentiment(texts, tokenizer, model):
    sentiments = np.zeros(3)  # Assuming the order [negative, neutral, positive]
    for text in texts:
        chunks = chunk_text(text, 512)  # 512 is the typical max length for transformers
        chunk_scores = np.zeros(3)
        for chunk in chunks:
            # Convert chunk to string and encode
            chunk = tokenizer.convert_tokens_to_string(chunk)
            encoded_input = tokenizer(chunk, return_tensors='pt', padding=True, truncation=True, max_length=512)
            # Predict sentiment
            with torch.no_grad():
                output = model(**encoded_input)
                scores = output.logits.detach().numpy()
                scores = softmax(scores, axis=1).mean(axis=0)
                chunk_scores += scores
        # Average the scores across chunks
        chunk_scores /= len(chunks)
        sentiments += chunk_scores
    # Normalize to get mean sentiment scores
    mean_sentiments = sentiments / len(texts)
    return mean_sentiments

labels = ['negative', 'neutral', 'positive']

# Display results: Specific feedback
story_continuations = [item["story_continuation"] for item in specific_feedback_modality_english]
ad_computer_mice = [item["ad_computer_mouse"] for item in specific_feedback_modality_english]
mean_sentiment_story_continuation = predict_sentiment(story_continuations, tokenizer, model)
mean_sentiment_ad_computer_mouse = predict_sentiment(ad_computer_mice, tokenizer, model)
print("Mean Sentiment - Story Continuation SPECIFIC:")
for label, score in zip(labels, mean_sentiment_story_continuation):
    print(f"{label}: {np.round(float(score), 4)}")
print("\nMean Sentiment - Ad Computer Mouse SPECIFIC:")
for label, score in zip(labels, mean_sentiment_ad_computer_mouse):
    print(f"{label}: {np.round(float(score), 4)}")

# Display results: General feedback
story_continuations = [item["story_continuation"] for item in general_feedback_modality_english]
ad_computer_mice = [item["ad_computer_mouse"] for item in general_feedback_modality_english]
mean_sentiment_story_continuation = predict_sentiment(story_continuations, tokenizer, model)
mean_sentiment_ad_computer_mouse = predict_sentiment(ad_computer_mice, tokenizer, model)
print("Mean Sentiment - Story Continuation GENERAL:")
for label, score in zip(labels, mean_sentiment_story_continuation):
    print(f"{label}: {np.round(float(score), 4)}")
print("\nMean Sentiment - Ad Computer Mouse GENERAL:")
for label, score in zip(labels, mean_sentiment_ad_computer_mouse):
    print(f"{label}: {np.round(float(score), 4)}")

# Display results: Control group
story_continuations = [item["story_continuation"] for item in control_group_english]
ad_computer_mice = [item["ad_computer_mouse"] for item in control_group_english]
mean_sentiment_story_continuation = predict_sentiment(story_continuations, tokenizer, model)
mean_sentiment_ad_computer_mouse = predict_sentiment(ad_computer_mice, tokenizer, model)
print("Mean Sentiment - Story Continuation CONTROL:")
for label, score in zip(labels, mean_sentiment_story_continuation):
    print(f"{label}: {np.round(float(score), 4)}")
print("\nMean Sentiment - Ad Computer Mouse CONTROL:")
for label, score in zip(labels, mean_sentiment_ad_computer_mouse):
    print(f"{label}: {np.round(float(score), 4)}")
```

```python
from transformers import pipeline, AutoTokenizer

# Initialize the sentiment analysis pipeline
sentiment_analysis = pipeline("sentiment-analysis", model="siebert/sentiment-roberta-large-english")
tokenizer = AutoTokenizer.from_pretrained("siebert/sentiment-roberta-large-english")

def chunk_text(tokenizer, text, max_length):
    tokens = tokenizer.tokenize(text)
    # Account for special tokens [CLS] and [SEP]
    max_chunk_size = max_length - 2
    chunks = [tokens[i:i + max_chunk_size] for i in range(0, len(tokens), max_chunk_size)]
    chunked_texts = [tokenizer.convert_tokens_to_string(chunk) for chunk in chunks]  # Corrected line
    return chunked_texts


def predict_sentiment_for_list(texts, tokenizer):
    sentiments_summary = {'POSITIVE': 0, 'NEGATIVE': 0, 'NEUTRAL': 0}
    num_texts = len(texts)
    
    for text in texts:
        # Prepare text in chunks
        chunked_texts = chunk_text(tokenizer, text, 512)  # 512 is a typical model's max sequence length
        
        for chunk in chunked_texts:
            # Perform sentiment analysis
            result = sentiment_analysis(chunk)[0]
            sentiments_summary[result['label']] += result['score']
    
    # Convert to average sentiment score
    for sentiment in sentiments_summary:
        sentiments_summary[sentiment] /= num_texts
    return sentiments_summary


# Extract texts
story_continuations = [item["story_continuation"] for item in specific_feedback_modality_english]
ad_computer_mice = [item["ad_computer_mouse"] for item in specific_feedback_modality_english]

# Calculate sentiments
mean_sentiment_story_continuation = predict_sentiment_for_list(story_continuations, tokenizer)
mean_sentiment_ad_computer_mouse = predict_sentiment_for_list(ad_computer_mice, tokenizer)

# Function to print results
def print_sentiment_scores(mean_sentiments, category_name):
    print(f"Mean Sentiment Scores for {category_name}:")
    for sentiment, score in mean_sentiments.items():
        print(f"{sentiment.title()}: {score:.4f}")
    print()

# Displaying the results
print_sentiment_scores(mean_sentiment_story_continuation, "Story Continuation SPECIFIC")
print_sentiment_scores(mean_sentiment_ad_computer_mouse, "Ad Computer Mouse SPECIFIC")

# general
story_continuations = [item["story_continuation"] for item in general_feedback_modality_english]
ad_computer_mice = [item["ad_computer_mouse"] for item in general_feedback_modality_english]

# Calculate sentiments
mean_sentiment_story_continuation = predict_sentiment_for_list(story_continuations, tokenizer)
mean_sentiment_ad_computer_mouse = predict_sentiment_for_list(ad_computer_mice, tokenizer)

# Displaying the results
print_sentiment_scores(mean_sentiment_story_continuation, "Story Continuation GENERAL")
print_sentiment_scores(mean_sentiment_ad_computer_mouse, "Ad Computer Mouse GENERAL")

# Calculate sentiments control
story_continuations = [item["story_continuation"] for item in control_group_english]
ad_computer_mice = [item["ad_computer_mouse"] for item in control_group_english]
mean_sentiment_story_continuation = predict_sentiment_for_list(story_continuations, tokenizer)
mean_sentiment_ad_computer_mouse = predict_sentiment_for_list(ad_computer_mice, tokenizer)
# Displaying the results
print_sentiment_scores(mean_sentiment_story_continuation, "Story Continuation CONTROL")
print_sentiment_scores(mean_sentiment_ad_computer_mouse, "Ad Computer Mouse CONTROL")
```

**Hypothesis 1: Specificity of Feedback and Creativity**

- **Alternative Hypothesis (H1a)**: The specificity of feedback provided by LLMs is positively correlated with the creativity of human work. That is, individuals receiving specific, detailed feedback on their creative tasks will produce work of higher creativity compared to those receiving generic feedback.
- **Null Hypothesis (H0a)**: There is no significant difference in the creativity of human work influenced by the specificity of feedback from LLMs. Specific and generic feedback lead to similar levels of creativity.

This hypothesis focuses on the core concept of the research question: whether detailed guidance from the LLM enhances creativity compared to generic feedback.

```python
# This Hypothesis is independent of the type of the creative task. Therefore I need to aggregate the data of BOTH creative tasks for each modality (general specific) OR conduct 2 independent t tests.
# For Hypothesis H1a (comparing specific feedback with general feedback):
from scipy import stats
from sentence_transformers import util
import torch
import pandas as pd
# import scipy.stats as stats

model = SentenceTransformer('sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2')

# Function to calculate similarities for each text within its group
def calculate_individual_similarities(texts, model):
    embeddings = model.encode(texts, convert_to_tensor=True)
    similarities = []
    for i in range(len(texts)):
        sim_scores = util.cos_sim(embeddings[i], embeddings)  # Compare one text to all others
        # Exclude self-comparison for calculating average similarity
        sim_scores = [sim_scores[0, j].item() for j in range(len(texts)) if i != j]
        similarities.extend(sim_scores)  # Collect all scores (excluding self-similarity)
    return similarities

# SBERT: Testing the Hypothesis with a independent two sampled t test for the story_continuations task 

specific_similarities_story = calculate_individual_similarities(story_continuations_specific, model)
general_similarities_story = calculate_individual_similarities(story_continuations_general, model)

specific_similarities_ad = calculate_individual_similarities(ad_computer_mice_specific, model)
general_similarities_ad = calculate_individual_similarities(ad_computer_mice_general, model)


# Perform t-tests to test hypothesis
ttest_result_story = stats.ttest_ind(specific_similarities_story, general_similarities_story, equal_var=False) # exact same task with distinct modalities
ttest_result_ad =stats.ttest_ind(specific_similarities_ad,general_similarities_ad, equal_var=False) # exact same task with distinct modalities
print(f"T-test result Specific vs General in Story Continuation: {ttest_result_story}")
print(f"T-test result Specific vs General in Advertisement: {ttest_result_ad}")
print("\n")


# Combine similarity scores across tasks for specific feedback
specific_similarities = []
specific_similarities.extend(specific_similarities_story)
specific_similarities.extend(specific_similarities_ad)

# Combine similarity scores across tasks for general feedback
general_similarities = []
general_similarities.extend(general_similarities_story)
general_similarities.extend(general_similarities_ad)

ttest_result_modality = stats.ttest_ind(specific_similarities, general_similarities, equal_var=False)
print(f"T-test result Specific vs General Modality: {ttest_result_modality}") # pvalue=0.0016882382876003258 -> reject the null hypothesis


# print(f"T-test result Specific Modality (Story continuation + Advertisement)", ) # both tasks in one modality compared to both tasks in the other modality.
# print(f"T-test result General Modality (Story continuation + Advertisement)", )
```

- **Null Hypothesis (H0d):** There is no significant difference in the average sentiment depth between creative outputs generated under the specific feedback condition and the general feedback condition
- **Alternative Hypothesis (H1d):** The average sentiment depth of creative outputs in the specific feedback condition is greater than the average sentiment depth in the general feedback condition, indicating more creative elements for the specific feedback condition

```python
# Calculate sentiments for every entry for task type and modality

import numpy as np
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from scipy.special import softmax

# Initialize tokenizer and model
MODEL = "cardiffnlp/twitter-roberta-base-sentiment-latest" # English
tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = AutoModelForSequenceClassification.from_pretrained(MODEL)

def preprocess(text):
    """Preprocess the text by replacing URLs and user mentions."""
    new_text = []
    for t in text.split(" "):
        t = '@user' if t.startswith('@') and len(t) > 1 else t
        t = 'http' if t.startswith('http') else t
        new_text.append(t)
    return " ".join(new_text)

def chunk_text(text, max_length):
    """Chunk text into segments that fit within the model's maximum input length."""
    tokens = tokenizer.tokenize(preprocess(text))
    num_chunks = len(tokens) // (max_length - 2) + 1
    chunks = [
        tokens[i * (max_length - 2): (i + 1) * (max_length - 2)]
        for i in range(num_chunks)
    ]
    return chunks

def predict_sentiment(texts, tokenizer, model):
    """Predict sentiment scores for a list of texts."""
    all_sentiments = []
    for text in texts:
        chunks = chunk_text(text, 512)
        chunk_scores = np.zeros(3)
        for chunk in chunks:
            chunk = tokenizer.convert_tokens_to_string(chunk)
            encoded_input = tokenizer(chunk, return_tensors='pt', padding=True, truncation=True, max_length=512)
            with torch.no_grad():
                output = model(**encoded_input)
                scores = output.logits.detach().numpy()
                scores = softmax(scores, axis=1).mean(axis=0)
                chunk_scores += scores
        chunk_scores /= len(chunks)
        all_sentiments.append(chunk_scores)
    return all_sentiments

labels = ['negative', 'neutral', 'positive']

# Gather data
data_sources = {
    "specific": {
        "story_continuations": [item["story_continuation"] for item in specific_feedback_modality_english],
        "ad_computer_mice": [item["ad_computer_mouse"] for item in specific_feedback_modality_english]
    },
    "general": {
        "story_continuations": [item["story_continuation"] for item in general_feedback_modality_english],
        "ad_computer_mice": [item["ad_computer_mouse"] for item in general_feedback_modality_english]
    },
    "control": {
        "story_continuations": [item["story_continuation"] for item in control_group_english],
        "ad_computer_mice": [item["ad_computer_mouse"] for item in control_group_english]
    }
}

# Predict sentiments and store results for t-test comparison
sentiment_results = {}
for group, contents in data_sources.items():
    sentiment_results[group] = {}
    for key, texts in contents.items():
        sentiment_results[group][key] = predict_sentiment(texts, tokenizer, model)

# Example: print all sentiments for specific feedback on story continuations
print("Sentiments for Specific Feedback - Story Continuations:")
for sentiments in sentiment_results["specific"]["story_continuations"]:
    print({label: np.round(float(score), 4) for label, score in zip(labels, sentiments)})
```

```python
# T test the sentiment depth
from scipy.stats import ttest_ind
import numpy as np

# Retrieve sentiment scores for specific and general feedback modalities
specific_sentiments_story = np.array(sentiment_results["specific"]["story_continuations"])
specific_sentiments_ad = np.array(sentiment_results["specific"]["ad_computer_mice"])
general_sentiments_story = np.array(sentiment_results["general"]["story_continuations"])
general_sentiments_ad = np.array(sentiment_results["general"]["ad_computer_mice"])

# Combine story and ad sentiments for each modality
specific_sentiments = np.vstack((specific_sentiments_story, specific_sentiments_ad))
general_sentiments = np.vstack((general_sentiments_story, general_sentiments_ad))

# Perform t-tests for each sentiment category
labels = ['negative', 'neutral', 'positive']
t_test_results = {}

for i, label in enumerate(labels):
    stat, p_value = ttest_ind(specific_sentiments[:, i], general_sentiments[:, i], equal_var=False)  # using Welch's t-test for unequal variances
    t_test_results[label] = (stat, p_value)

# Display the results
for label, results in t_test_results.items():
    print(f"T-test results for {label} sentiment:")
    print(f"Statistic: {results[0]}, P-value: {results[1]}")
```

# Measuring Text Quality
There's no single perfect metric but we will use some recognized measures in order to measure factors like readability

```python
from textstat import flesch_kincaid_grade # Readability
# This metric estimates the U.S. education level needed to understand the text. A lower score indicates easier readability. 

def calculate_flesch_kincaid_grade(text):
    """
    Calculates the Flesch-Kincaid readability score for a given text.

    Args:
        text: A string containing the text to analyze.

    Returns:
        A float representing the Flesch-Kincaid score for the text.
    """
    try:
        grade = flesch_kincaid_grade(text)
    except:
        grade = 0.0  # Handle exceptions (e.g., empty text)
    return grade


def coleman_liau_index(text):
    """
    Calculates the Coleman-Liau Index (CLI) for a given text.

    Args:
        text: A string containing the text to analyze.

    Returns:
        A float representing the CLI score of the text.
    """
    import re

    # Count the number of letters
    char_count = sum(c.isalpha() for c in text)
    
    # Count the number of words
    words = len(text.split())
    
    # Count the number of sentences
    sentences = len(re.split(r'[.!?]+', text)) - 1

    # Calculate the average number of letters per 100 words (L)
    L = (char_count / words * 100) if words > 0 else 0

    # Calculate the average number of sentences per 100 words (S)
    S = (sentences / words * 100) if words > 0 else 0

    # Calculate the Coleman-Liau Index
    CLI = 0.0588 * L - 0.296 * S - 15.8
    return CLI

flesch_scores_specific = []
flesch_scores_general = []
flesch_scores_control = []

cli_specific = []
cli_general = []
cli_control = []


for text in story_continuations_specific:
    score = calculate_flesch_kincaid_grade(text)
    cli_score = coleman_liau_index(text)
    flesch_scores_specific.append(score)
    cli_specific.append(cli_score)

for text in story_continuations_general:
    score = calculate_flesch_kincaid_grade(text)
    cli_score = coleman_liau_index(text)
    flesch_scores_general.append(score)
    cli_general.append(cli_score)

for text in story_continuations_control:
    score = calculate_flesch_kincaid_grade(text)
    cli_score = coleman_liau_index(text)
    flesch_scores_control.append(score)
    cli_control.append(cli_score)


# Calculate average scores for each list
specific_flesch_average_score = sum(flesch_scores_specific) / len(flesch_scores_specific)
genereal_flesch_average_score = sum(flesch_scores_general) / len(flesch_scores_general)
control_flesch_average_score = sum(flesch_scores_control) / len(flesch_scores_control)

specific_cli_average_score = sum(cli_specific) / len(cli_specific)
general_cli_average_score = sum(cli_general) / len(cli_general)
control_cli_average_score = sum(cli_control) / len(cli_control)

# Print the results
print("Average Flesch-Kincaid Score for Specific Story Continuations:", specific_flesch_average_score)
print("Average Flesch-Kincaid Score for General Story Continuations:", genereal_flesch_average_score)
print("Average Flesch-Kincaid Score for Control Story Continuations:", control_flesch_average_score)
print("\n")
print("Average CLI score for Specific Story Continuations:", specific_cli_average_score)
print("Average CLI score for General Story Continuations:", general_cli_average_score)
print("Average CLI score for Control Story Continuations:", control_cli_average_score)


# You can now compare the average scores to see which list has the most readable text.
```
